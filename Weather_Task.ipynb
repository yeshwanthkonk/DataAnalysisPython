### <font color = green>Adding Mean or Median to Columns which are having NaN(means zeros)</font>

```python
mn = df['Temperature (C)'].mean()
md = df['Temperature (C)'].median()
df.loc[np.isnan(df['Temperature (C)']),'Temperature (C)'] = mn if int(mn) == int(md) else md
```

```python
mn = df['Wind Speed (km/h)'].mean()
md = df['Wind Speed (km/h)'].median()
df.loc[np.isnan(df['Wind Speed (km/h)']),'Wind Speed (km/h)'] = mn if int(mn) == int(md) else md
```

```python
mn = df['Wind Bearing (degrees)'].mean()
md = df['Wind Bearing (degrees)'].median()
df.loc[np.isnan(df['Wind Bearing (degrees)']),'Wind Bearing (degrees)'] = mn if int(mn) == int(md) else md
```

```python
mn = df['Pressure (millibars)'].mean()
md = df['Pressure (millibars)'].median()
df.loc[np.isnan(df['Pressure (millibars)']),'Pressure (millibars)'] = mn if int(mn) == int(md) else md
```

```python
df.info()
```

#### Values which are not raining hours, in data set is 1230

### <font color=red>Formatted Date, Daily Summary Column is not much usefull for our analysis, So we remove that columns</font>

```python
del df['Formatted Date']
del df['Daily Summary']
```

##### Droping remaining 9 columns doesn't effect Analysis, So remove these rows

```python
df.dropna(inplace = True)
```

```python
df.info()
```

### <font color = red> Removing Duplicate rows, if available </font>

```python
print("Before : ",df.shape)
df.drop_duplicates(inplace = True)
print("After : ",df.shape)
```
